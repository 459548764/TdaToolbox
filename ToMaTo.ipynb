## Concept

Here follow my implementation of the ToMaTo clustering algorithm, presented by Steve Oudot in its book "Persistence Theory: From Quiver Representations to Data Analysis". It mainly consists in exploiting the strength of topological data analysis for clustering. A few examples similar to the ones provided by scikit-learn are given below. The idea behind the algorithm is the following: Given a two-dimensional dataset, the sublevels filtration of the simplextree build over the density estimate of the the data gives, by reconstruction of the graph of the connected components, the hierarchical relation between the data points. By sorting it, we are able to retrace the density extrema, and thus the centroids. Visualized persistence, as presented below, gives a good indication over the number of centroids. Once determined, the construction of the distinct sets through UnionFind becomes then possible.

## Examples

```python
gen = ClusterGenerator(structure='anisotropy', randomize=45)
clf = ToMaTo(*gen.generate())
_ = clf.estimate_clusters(graph=True)
```

```python
gen = ClusterGenerator(structure='anisotropy', randomize=45)
clf = ToMaTo(*gen.generate())
_ = clf.estimate_density(graph=True)
clf.estimate_clusters(graph=False)
prd = clf.fit_predict(num_clusters=3, graph=True)
```
